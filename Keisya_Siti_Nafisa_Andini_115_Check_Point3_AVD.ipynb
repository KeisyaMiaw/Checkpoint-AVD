# Membaca Data

```python
df = pd.read_csv("/content/smartphones_cleaned_v6_new.csv")

df.head()
```

# Visualisasi Data

# Bar Chart

```python
print("Menampilkan Bar Chart: Menghitung Akumulasi Jumlah Rating per Brand")
plt.figure(figsize=(23, 10))
df.groupby('brand_name')['rating'].sum().plot(kind='bar', color='blue')
plt.title('Jumlah Rating per Brand')
plt.xlabel('Nama Brand')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

# Insight Data dan Tindakan Bar Chart

**Insight Data**
1. Rating tertinggi didapatkan oleh dengan nama brand xiaomi, samsung, dan vivo. Dikarenakan harga yang lebih terjangkau dan fitur yang dibutuhkan oleh para konsumen.
2. Sedangkan, untuk rating menengah didapatkan oleh dengan nama brand realme, oppo, dan apple.
3. Adapun nama brand dengan rating terendah, ada yang hingga mencapai nilai 0, kemungkinan dikarenakan nama brand yang kurang terkenal dan juga kurang diminati dikalangan kostumer.

**Tindakan**
1. Bagi brand yang mendapatkan rating tertinggi harus tetap mengoptimalkan pemasaran dan promosi.
2. Lalu, bagi brand yang mendapatkan rating menengah bisa menyesuaikan pada brand yang mendapatkan rating tertinggi.
3. Untuk brand mendapatkan rating paling rendah, bisa meneliti mengapa brand tersebut kurang terkenal dan kurang diminati oleh para kostumer.

# Pie Chart

```python
print("Menampilkan Pie Chart: Menghitung Konstribusi OS per Harga ")
plt.figure(figsize=(10, 10))
df.groupby('os')['price'].sum().plot(kind='pie', autopct='%1.1f%%', colors = ['#FF5733', '#33FF57', '#3357FF', '#F1C40F'])
plt.title('Jumlah Konstribusi OS per Harga')
plt.ylabel('')
plt.show()
```

# Insight Data dan Tindakan Pie Chart

**Insight Data**
1. Android memiliki nilai konstribusi lebih tinggi yaitu 83,2% yang menunjukkan bahwa mayoritas smartphone memiliki *Operating System* ini. Dikarenakan harganya yang lebih terjangkau.
2. iOS berada pada urutan kedua yaitu memiliki nilai 14,4%, karena iOS memberikan fitur yang lebih canggih dan premium, namun disayangkan harganya yang cukup tinggi.

**Tindakan**
1. Untuk android, harus lebih memperkuat pemasaran dan juga peluang yang besar meningkatkan pengalaman premium supaya bisa kompetitif dengan iOS.
2. Untuk iOS, bisa mengadakan promo dengan harga yang terjangkau untuk pengguna baru apple, agar tertarik dengan brand tersebut. Bisa juga mempertahankan perbedaan eksleusif untuk tetap menarik pengguna lama.


# Boxplot

```python
print("Menampilkan Boxplot: Menghitung Distribusi Refresh Rate Berdasarkan Nama Brand:")
plt.figure(figsize=(27, 10))
sns.boxplot(x='brand_name', y='refresh_rate',hue="brand_name", data=df, palette='mako')
plt.title('Dsitribusi Rating Refresh berdasarkan Nama Brand')
plt.xlabel('Nama Brand')
plt.ylabel('Rate')
plt.xticks(rotation=45)
plt.show()
```

# Insight Data dan Tindakan Boxplot

**Insight Data**
1. Ada beberapa brand yang menawarkan refresh rate standar, namun ada juga yang menawarkan refresh rate yang ekstrem.
2. Brand yang menawarkan refresh rate yang konsisten seperti samsung, apple, xiaomi dan oppo.

**Tindakan**
1. Untuk bagi produsen, brand yang memiliki refresh rate yang ekstrem bisa menargetkan pasar pada gamer. Sedangkan, untuk refresh rate yang lebih rendah bisa mempertimbangkan untuk meningkatkan refresh rate, dikarenakan ini telah digunakan oleh mayoritas brand.
2. Untuk bagi para konsumen, jika ingin mencari yang pengalaman visual nya lebih bagus dan lancar bisa memilih yang refresh rate lebih tinggi. Namun, jika mencari yang hanya digunakan untuk sehari-hari saja, maka pilih brand yang memilii refresh rate standar/lebih rendah.
